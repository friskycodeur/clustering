## <center> K-Means Clutering on Customer Data <center> 

In this project , we will use a dataset of customers of a mall consisting of their details such as Annual Income , spending score(given as per the spendings they have in the mall), age , etc.
We will use this data to observe their spendings , categoize them using k-means clustering and then at the end give some inputs so as to how we can influence different category of these customers to 
help our mall grow.
The necessary dataset is already attahed within the repository.

***

### Importing the libraries

In this project we will need pandas and numpy for playing with the data, and we will visualize the data and observe our customers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

***

### Reading the data in CSV file !

As we don't have any depedent variables in the dataset , we will not assign y any values.
And here we are taking only Salary and Spending score in consideration to visualize and observe the customers.

```python
df=pd.read_csv('Mall_Customers.csv')
X=df.iloc[:,[3,4]].values
```

***

### Using the elow method to get the optimal number of clusters

Now let us use the elbow method to see what is the optimal number of clusters that we should use.
(If you don't know about the elbow method here's a link for you : [Elbow Method](https://www.geeksforgeeks.org/elbow-method-for-optimal-value-of-k-in-kmeans/))

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans =KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('Wcss')
```

<br> Clearly from the plot , the optimal number of clusters is 5. So let us now build a K-means model with 5 clusters and categorize the customers under these five clusters.

```python
kmeans =KMeans(n_clusters=5,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(X)
print(y_kmeans)
```

***

### Visualizing the Clusters

Now let's visualize the clusters we have just created and see how the customers are being categorized. And we can also take some 
tips on what we should do to get more from out customers. Where to spend in advertising , where to advertize more and where to advertize less, etc.
So let's jump into it ! 

```python
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=100,c='green',label='Cluster 2')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=100,c='blue',label='Cluster 3')
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],s=100,c='cyan',label='Cluster 4')
plt.scatter(X[y_kmeans==4,0],X[y_kmeans==4,1],s=100,c='violet',label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='centroids')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income of Customers')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
