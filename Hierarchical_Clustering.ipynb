# <center> Hierarchical Clustering on Customers data </center>

In this project , we will use a dataset of customers of a mall consisting of their details such as Annual Income , spending score(given as per the spendings they have in the mall), age , etc.
We will use this data to observe their spendings , categoize them using means hierarchical clustering and then at the end give some inputs so as to how we can influence different category of these customers to 
help our mall grow.
The necessary dataset is already attahed within the repository.

***

### Importing the Libraries

In this project we will need pandas and numpy for playing with the data, and we will visualize the data and observe our customers.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

***

### Reading the CSV file

As we don't have any depedent variables in the dataset , we will not assign y any values.
And here we are taking only Salary and Spending score in consideration to visualize and observe the customers.

```python
df=pd.read_csv('Mall_Customers.csv')
X=df.iloc[:,[3,4]].values
```

***

### Using the Dendogram to get the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendogram= sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendogram')
plt.xlabel('Number of Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

As clearly the longest vertical line (with no interfering horizontal line in between) is the <font color='blue'> blue one </font> on the most right, and thus we will get the optimal number of clusters to be 5.

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(X)
```

***

### Visualizing the Clusters

Now let's visualize the clusters we have just created and see how the customers are being categorized. And we can also take some 
tips on what we should do to get more from out customers. Where to spend in advertising , where to advertize more and where to advertize less, etc.
So let's jump into it ! 

```python
plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=100,c='green',label='Cluster 2')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=100,c='blue',label='Cluster 3')
plt.scatter(X[y_hc==3,0],X[y_hc==3,1],s=100,c='cyan',label='Cluster 4')
plt.scatter(X[y_hc==4,0],X[y_hc==4,1],s=100,c='violet',label='Cluster 5')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income of Customers')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
